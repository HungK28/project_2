```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
imdb_file = "Resources/IMDb movies.csv"
imdb_df = pd.read_csv(imdb_file)
imdb_df.head()
```

```python
cleaned_imdb_df = imdb_df[['title','avg_vote']].copy()
cleaned_imdb_df.head()
```

```python
netflix_file = "Resources/netflix_titles.csv"
netflix_df = pd.read_csv(netflix_file)
netflix_df.head()
```

```python
cleaned_netflix = netflix_df[['title', 'director','release_year', 'country']].copy()
cleaned_netflix.head()
```

```python
new_netflix = cleaned_netflix.rename(columns = {'release_year': 'year'})
new_netflix.head()
```

```python
new_netflix.drop_duplicates(subset=['title'], inplace = True)
new_netflix.dropna(subset=['title'], inplace = True)
new_netflix.set_index('title', inplace = True)
```

```python
new_netflix = new_netflix.reset_index()
```

```python
new_netflix
```

```python
cleaned_imdb_df.drop_duplicates(subset=['title'], inplace = True)
cleaned_imdb_df.dropna(subset=['title'], inplace = True)
cleaned_imdb_df.set_index('title', inplace = True)
```

```python
cleaned_imdb_df
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/movies_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
cleaned_imdb_df.to_sql(name='imdb', con=engine, if_exists='append', index=True)
```

```python
new_netflix.to_sql(name='netflix', con=engine, if_exists='append', index=True)
```

```python
engine.table_names()
```

```python
pd.read_sql_query('select cn.title, cn.director, cn.year, cn.country, avg_vote from netflix cn join imdb cl on cn.title=cl.title', con=engine)
```
